```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%sql select * from SPACEXTABLE limit 50
```

```python
%sql select Launch_Site from SPACEXTABLE group by Launch_Site
```

```python
%sql select * from SPACEXTABLE where Launch_Site like 'CCA%' limit 5
```

```python
%sql select sum(PAYLOAD_MASS__KG_) from SPACEXTABLE where Customer like 'NASA (CRS)%'
```

```python
%sql select avg(PAYLOAD_MASS__KG_) from SPACEXTABLE where Booster_Version like 'F9 v1.1%'
```

```python
%sql select min(Date) from SPACEXTABLE 
```

```python
%sql select Booster_Version from SPACEXTABLE where Landing_Outcome = 'Success (ground pad)' and 4000<PAYLOAD_MASS__KG_<6000
```

```python
%sql select Mission_Outcome,count(Mission_Outcome) from SPACEXTABLE group by Mission_Outcome
```
